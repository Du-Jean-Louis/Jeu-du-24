```python
import numpy as np

def operation2nombres(a, b):
    """Renvoie toute les combinaisons possibles"""  
    res = []
    
    # Ajout addition
    res.append(a+b)
    
    # Ajout soustraction
    res.append(a-b)
    res.append(b-a)

    # Ajout multiplication
    res.append(a*b)

    # Ajout division
    if b!=0:
        res.append(a/b)
    
    if a!=0:
        res.append(b/a)

    # On met pas encore les puissances

    return res

operation2nombres(3,4)
        
```

```python
import numpy as np

def somme_produit_2nombres(a, b):
    """Renvoie toute les combinaisons possibles"""  
    res = []
    
    # Ajout addition
    res.append(a+b)

    # Ajout multiplication
    res.append(a*b)

    return res

somme_produit_2nombres(3,4)
        
```

```python
def operation2nombres_plus(a, b):
    """Renvoie toute les combinaisons possibles
    + Garder en mémoire les opérations faites"""  
    res = {}
    
    # Ajout addition

    res[f"{a}+{b}"] = a+b
    
    # Ajout soustraction
    res[f"{a}-{b}"] = a-b
    res[f"{b}-{a}"] = b-a

    # Ajout multiplication
    res[f"{a}*{b}"] = a*b

    # Ajout division
    if b!=0:
        res[f"{a}/{b}"] = a/b
    
    if a!=0:
        res[f"{b}/{a}"] = b/a

    # On met pas encore les puissances

    return res

operation2nombres_plus(3,4)
```

```python
# Comment doit fonctionner l'algorithme ? 
# Il doit parcourir chaque paires

def parcours_paires(list):
    """Parcours chaque paires"""
    for i in range(len(list)):
        for j in range(i+1,len(list)):
            print(f"({list[i]},{list[j]})")

            for k in range(len(list)):
                if k!=i and k!=j:
                    print(list[k])    

parcours_paires([1,2,3,4])
```

```python
def f(list):

    print("debut_fonction",list)
    n = len(list)
    # Sortie de la récursion
    if n==1:
        if list[0] == 24:
            return True
    
    for i in range(n):
        for j in range(i+1,n):
            print("__________________")
            l = operation2nombres(list[i],list[j])
            
            list_k = []
            for k in range(n):
                if k!=i and k!=j:
                    list_k.append(list[k])

            print(f"list_k : {list_k}")
            #print(f"l : {l}")
            for r in l:
                new_list = list_k.copy()
                new_list.append(r)
                print(f"list_k : {list_k}, l : {l}, new_list : {new_list}")
                if f(new_list):
                    return True

    return False

f([10,2,9,4])
```

```python
def f2(list,commentaire=""):

    #print("debut_fonction",list)
    n = len(list)
    # Sortie de la récursion
    if n==1:
        # Problèmes certaines solutions fractionnelles, ne tombent pas sur 24 pile
        if np.abs(list[0] - 24)<0.0001:
            print(commentaire)
            return True
        else :
            return False
    
    for i in range(n):
        for j in range(i+1,n):
            #print("__________________")
            # l est un dico
            l = operation2nombres_plus(list[i],list[j])
            
            list_k = []
            for k in range(n):
                if k!=i and k!=j:
                    list_k.append(list[k])

            #print(f"list_k : {list_k}")
            #print(f"l : {l}")
            for comm, r in l.items():
                new_list = list_k.copy()
                new_list.append(r)
                #print(f"list_k : {list_k}, l : {l}, new_list : {new_list}")
                if f2(new_list, commentaire+"\n"+comm):
                    return True

    return False

f2([3,8,3,8])
```
